```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y, 'r--')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plot")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'g')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# OO

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title("LARGER PLOT")
axes2.plot(y, x)
axes2.plot("SMALLER PLOT")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title("FIRST")

axes[1].plot(y,x)
axes[1].set_title("FIRST")

plt.tight_layout()
```

## FIGURE SIZE AND DPI

```python
fig = plt.figure(figsize=(8, 2))

ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.plot(x, y)
```

```python
fig.savefig('image.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax.plot(x, x ** 2, label="Squared")
ax.plot(x, x ** 3, label="Cubed")
ax.legend(loc=10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="#ff8c00", linewidth=3, linestyle='--', alpha=0.5, marker='o')
```
